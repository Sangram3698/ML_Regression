```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
```

```python
dataset
```

```python
X=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
#training the linear regression model on the whole dataset
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X,y)
```

```python
#training the polynomial regression model
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=6)
X_poly=poly.fit_transform(X)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_poly,y)
```

```python
#visualising the linear regression model
plt.scatter(X,y,color='red')
plt.plot(X,regressor.predict(X),color='blue')
plt.title('Through linear regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
#visualising the polynomial regression model
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg.predict(poly.fit_transform(X)),color='black')
plt.title('Through polynomial regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
#for more smooth curve
X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,color='red')
plt.plot(X_grid,lin_reg.predict(poly.fit_transform(X_grid)),color='blue')
plt.title('Through polynomial regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
